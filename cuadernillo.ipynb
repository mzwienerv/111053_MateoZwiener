```python
# %% EJERCICIO 2 LEER UN NUMERO ENTERO Y ESCRIBIRLO EN REVERSA


num = int(input('Ingrese un número entero: '))

##EL PROGRAMA DECIDE SI EL NUMERO ES POSITIVO O NEGATIVO 

if num < 0: 
    signo = -1
else:
    signo = 1
    
num = abs(num)  
inv = 0
###AQUI PROCEDE A INVERTIR CADA NUMERO MIENTRAS SE CUMPLE EL CICLO DEL TAMAÑO
while num/10 > 0:    
    n = num % 10    
    num = int(num/10) 
    inv = (inv*10) + n 

inv = inv*signo 
print('El inverso del numero es: ',inv)
```

```python
#%%EJERCICIO 5 TALLER Leer un número entero y determinar
##el número de cincos que están consecutivos en un número.


num = int(input('Escriba un número entero: '))                           
num2 = abs(num) 


##CON ESTA LISTA GUARDAREMOS LOS NUMEROS INGRESADOS
lista1 = []   

while num2 > 0:
    lista1.append(num2 %10)
    num2 //=10

lista1.append(0)  

##ESTE CICLO EMPEZARA A CONTAR LOS 5s SEGUIDOS QUE SE ENCUENTREN EN EL NUMERO
contador = 0         
for i in range(len(lista1)):
    if lista1[i] == 5:
        if lista1[i-1] == 5 or lista1[i+1] == 5:
            contador +=1

print(f'Hay {contador} cincos consecutivos en el número {num}')
```

```python
#%%EJERCICIO 6 TALLER Leer un número entero y determinar
## si la sumaf de sus dígitos es un número de Fibonacci.

##CREAMOS UNA FUNCION  QUE ACTUARA COMO LA SECUENCIA DE FIBONACCI
def fib(n):                            
    fibonacci = []
    a,b = 1,1
    while a<=n:
        fibonacci.append(a)
        a,b = b,b+a
    return fibonacci

num = int(input('Ingrese un número: '))
sumaf = 0

##CON ESTE CICLO EMPEZAMOS A SUMAR EL NUMERO INGRESADO PARA VER SI HACE PARTE
##DE LA SECUENCIA DE FIBONACCI

while num/10 > 0:
    n = num % 10
    num = int(num/10)
    sumaf = sumaf + n
    
if sumaf in fib(sumaf):
    print('La sumaf de los dígitos del número ingresado es un número de \
          fibonacci.')
else:
    print('La sumaf del número ingresado no es un número de fibonnaci.')
```

```python
#%% EJERCICIO 8 TALLER Iterar a través de los primeros cien enteros positivos, 
##buscando los múltiplos de 3 e imprimiendo y almacenándolos en una lista hasta
## encontrar los primeros 15 de ellos. Una vez encontrados, continuar iterando
## en busca de los múltiplos de 4 y almacenarlos en otra lista.
#%%Variables

rango_inferior = 1
rango_superior = 101
Limite_3 = 16
Inicio_4 = Limite_3-5
multiplo3=[3]  #Lista de los multiplos de 3
multiplo4=[]  #Lista de los multiplos de 4

#%% Función que servira para definir los multiplos que se usaran y buscaran

def multiplo(valor, multiplo):
    
    res = valor % multiplo
    if res == 0:
        return True
    else:
        return False

#%%Multiplos de 3
##EN ESTE CICLO BUSCAREMOS LOS MULTIPLOS DE 3 Y LOS AGREGAMOS A LA LISTA 
for i in range(rango_inferior,rango_superior):
    if multiplo(i,3):   
        multiplo3.append(i)   
print('Los primeros 15 multiplos de 3 son: ',multiplo3[rango_inferior:\
                                                         Limite_3])

#%%Multiplos de 4
## EN ESTE CICLO BUSCAREMOS LOS MULTIPLOS DE 4 PARA AGREGARLOS
for j in range(rango_inferior,rango_superior): 
    if multiplo(j,4):   
        multiplo4.append(j) 
print('De los restantes números, los multiplos de 4 son: ',multiplo4\
      [Inicio_4:rango_superior]) 
```

```python
#%% EJERCICIO 9 TALLER Este programa permite leer las coordenadas (x1, y1, r1)
## y (x2, y2, r2) que corresponden al centro y al radio de dos círculos. 
##Leer un punto de coordenadas (a, b). Y determina si (a, b) está contenido: a)
##dentro del círculo 1; b) dentro del círculo 2; c) dentro de ambos círculos. 
##Teniendo en cuenta que el círculo es el lugar geométrico de los puntos del 
##plano cuya distancia a otro punto fijo, llamado centro, es menor o igual que 
##una cantidad constante, llamada radio.

def distancia(x, y, a, b):
#la función perimite calcular la ditancia de dos puntos, fundamentada por el
#teorema de pitágoras
    dist = (((x - a) ** 2)+((y - b) ** 2)) ** (1 / 2)
    return dist

#%% Datos del círculo 1.
    
while True :
    try:
        coord_x_1 = int(input('Ingresar la coordenada en X del círculo 1: '))
        break
    except ValueError :
        print ('El valor ingresado no es correcto, intentelo de nuevo.')

while True :
    try:
        coord_y_1 = int(input('Ingresar la coordenada en Y del círculo 1: '))
        break
    except ValueError :
        print ('El valor ingresado no es correcto, intentelo de nuevo.')
        
#los ciclos while para las coordenadas del circulo 1 se usan para admitir uni-
#camente variables enteras, de manera que da la opción al usuario que si no lo
#hace, las ingrese nuevamente y de manera correcta.
        
while True :
    try:
        r_1 = int(input('Ingrese el radio del círculo 1: '))
        if r_1 >= 0 : break
    except ValueError :
        print ('El valor ingresado no es correcto, intente de nuevo.')
#El while para el radio tiene como misma función acpetar unicamente enteros

#%% Datos del circulo 2.
    
while True :
    try:
        coord_x_2 = int(input('Ingresar la coordenada en X del círculo 2: '))
        break
    except ValueError :
        print ('El valor ingresado no es correcto, intente de nuevo.')

while True :
    try:
        coord_y_2 = int(input('Ingresar la coordenada en Y del círculo 2: '))
        break
    except ValueError :
        print ('El valor ingresado no es correcto, intente de nuevo.')
        
while True :
    try:
        r_2 = int(input('Por favor ingrese el radio del círculo 2: '))
        if r_2 >= 0 : break
    except ValueError :
        print ('\nEl valor ingresado no es correcto, intente de nuevo')

#%% Ingresar las coordenadas del punto (a,b) a evaluar.
    
while True :
    try:
        a = int(input('Por favor ingrese la coordenada en X del punto: '))
        break
    except ValueError :
        print ('El valor ingresado no es correcto, intente de nuevo.')

while True :
    try:
        b = int(input('Por favor ingrese la coordenada en Y del punto: '))
        break
    except ValueError :
        print ('El valor ingresado no es correcto, intente de nuevo.')

#%% Ejecutar la función nombrada al inicio de código.
        
distancia_a_1 = distancia(coord_x_1, coord_y_1, a, b)
distancia_a_2 = distancia(coord_x_2, coord_y_2, a, b)

#%% Evaluar las condiciones de impresión.

if distancia_a_1 <= r_1 and distancia_a_2 <= r_2:
    print ('El punto', a,',', b, 'se encuentra dentro de ambos círculos.')
    
elif distancia_a_1 > r_1 and distancia_a_2 <= r_2:
    print ('El punto', a,',', b, 'se encuentra solo dentro del círculo 2.')
    
elif distancia_a_1 <= r_1 and distancia_a_2 > r_2:
    print ('El punto', a,',', b, 'se encuentra solo dentro del círculo 1.')
    
elif distancia_a_1 > r_1 and distancia_a_2 > r_2:
    print ('El punto', a,',', b, 'no se encuentra dentro de ningún círculo.')
```

```python
#%% EJERCICIO 11 TALLER Escribir un programa que reciba una cadena de texto 
##y reporte:
##a) cuántas letras vocales en mayúscula se entraron;
##b) cuántas letras con tilde se entraron (minúsculas y mayúsculas);
##c) cuántos dígitos se entraron,
##d) cuántos espacios se entraron;
##e) cuántas palabras reservadas se entraron.



#%%IMPORTAMOS LIBRERIA DE PALABRAS RESERVADAS
import keyword  
cadena = input('Ingrese una cadena de texto: ')

#%% Vocales en mayúsculas

contador = 0
voc_mayus=['A','E','I','O','U','Á','É','Í','Ó','Ú','Ü']

for x in cadena:
    if x in voc_mayus and cadena:
        contador += 1
print('La cantidad de vocales en mayusculas son: ',contador)

#%% Letras con tilde

contador2 = 0
voc_tildes = ['á','é','í','ó','ú','Á','É','Í','Ó','Ú']

for x in cadena:
    if x in voc_tildes and cadena:
        contador2 += 1
print('La cantidad de tildes son: ',contador2)

#%% Cantidad de dígitos

contador3 = 0

for x in cadena:
    if x.isdigit():
        contador3 += 1
print('La cantidad de dígitos que tiene la cadena son: ', contador3)

#%% Cantidad de espacio

contador4 = 0
espacio = [' ']

for x in cadena:
    if x in espacio and cadena:
        contador4 += 1
print('La cantidad de espacios son: ', contador4)

#%% Palabra reservadas

contador5 = 0
reservadas = keyword.kwlist
cad = cadena.split()

for x in cad:
    if x in reservadas and cad:
        contador5 += 1
print('La cantidad de palabras reservadas son: ',contador5)
```

```python
#%% EJERCICIO 13 Escribir un programa que lea una cadena de texto y la imprima 
##como un triángulo. Por ejemplo, si la cadena entrada es: "Es una locura odiar 
##a todas las rosas porque una te pinchó: renunciar a todos tus sueños porque 
##uno de ellos no se realizó - El Principito", el programa debe imprimir:
##            E
##           sun
##          alocu
##         raodiar
##        atodaslas
##       rosasporque
##      unatepinchó:r
##     enunciaratodost
##    ussueñosporqueuno
##   deellosnoserealizó-
##      ElPrincipito

cadena = input('Digite la cadena de texto: ')

# Se eliminan los espacios presentes en la cadena
cadena = cadena.replace(' ', '')

# Se determina el número de caracteres en la cadena
num_caract = len(cadena)

#%% Se determina el número de filas y el número máximo de caracteres por fila
temp = 0                       # Variable auxiliar 
num_filas = 1                  
while temp < num_caract:
    temp += 2 * num_filas - 1
    num_filas += 1

num_max = 2 * num_filas - 1    # Número máximo de caracteres en una fila

#%% Se imprime la cadena de texto 
aux_1 = 0        #Almacena la última posición en la que se imprime la fila i
aux_2 = 0        #Cantidad de letras que se imprimen en la fila i
contador = 0     # Verifica si luego de imprimir el triángulo aún quedan letras
                 # por imprimir

for i in range(1,num_filas):
    aux_2 += 2*i-1     
    print(cadena[aux_1:aux_2].center(num_max,' '))
    aux_1 = aux_2      
    contador += 1
    
if contador != num_caract:
    print(cadena[aux_2:].center(num_max,' '))
```
