```python
#%%% EJERCICIO 16
'''Leer una lista de números y decir si estan ordenados ascendentemente, 
   y luego pedir otro número y ordenarlo en la lista ascendente'''
    
n = int(input('¿Cuántos números desea ingresar? ')) #Pido el rango de la lista

lis_vacia=[]            #Creo una lista vacia

for i in range(n):
    a=int(input('Ingrese el número: '))   #Pido los numeros deacuerdo al rango
    lis_vacia.append(a)         #Agrego cada numero a la lista vacia
print('La lista de números ingresada es: ',lis_vacia)

lista = lis_vacia[:]        #Hago una copia de la lista
lis_vacia.sort()            #Ordeno la copia
if (lista == lis_vacia):    #Comparo la copia y el original
    print('La lista está correctamente ordenada.')
else:
    print('La lista no está ordenada correctamente.')

#%% Agregar otro número    
numero = int(input('Ingrese un número: '))

lis_vacia.append(numero) #le sumo el numero a la lista
lis_vacia.sort()        #Ordeno la nueva lista
print('La nueva lista ordenada es: ',lis_vacia)
```

```python
#%% EJERCICIO 18
'''
Este programa lee una matriz cuadrada A, obtiene su matriz triangular superior 
asociada U y determina qué fila de U contiene la
mayor cantidad de unos.
'''
import numpy as np
import copy

a = int(input('Ingrese el número de filas de su matriz: '))
b = int(input('Ingrese el número de columnas de su matriz: '))

A = np.array([[0 for x in range(b)]for y in range(a)])
for i in range(a):
    for j in range(b):
        A[i][j] = int(input(f'\nIngrese el número de la fila {i+1},'\
         f' columna {j+1}: '.format(i,j)))

#%% MATRIZ TRIANGULAR SUPERIOR 
U = copy.deepcopy(A)
for i in range(a):
    for j in range(b):
        if (i>j):
            U[i][j] = 0
print('\nLa matriz triangular superior de la matriz dada es: ')
print('\n',U)

#%% FILA CON MAYOR CANTIDAD DE UN NÚMERO
numero = int(input('Ingrese el número a identificar en la matriz: '))

U_list = U.tolist()
listas = [[] for i in range(a)]
for i in range(0,a):
    x=U_list[i].count(numero)
    listas[i].append(x)
    

    
print(f'La mayor cantidad de numeros {numero} en una fila es de: ',max(listas))
```

```python
#%% EJERCICIO 20
'''
Este programa lee una matriz y la rota 90 grados en sentido antihorario.
'''
import numpy as np

a = int(input('Ingrese el número de filas de su matriz: '))
b = int(input('Ingrese el número de columnas de su matriz: '))

A = np.array([[0 for x in range(b)]for y in range(a)])
for i in range(a):
    for j in range(b):
        A[i][j] = int(input(f'\nIngrese el número de la fila {i+1},'\
         f' columna {j+1}: '.format(i,j)))

#%% ROTAR MATRIZ SIN NUMPY
B = np.array([[A[j][i] for j in range(len(A))]\
                for i in range(len(A[0])-1,-1,-1)])

print('\nLa matriz ingresadad es: ')
print(A)
print('\nLa matriz rotada 90 grados en sentido antihorario es: ')
print(B)

#%% ROTAR MATRIZ CON NUMPY    
U = np.rot90(A)
print('\nLa matriz ingresadad es: ')
print(A)
print('\nLa matriz rotada 90 grados en sentido antihorario es: ')
print(U)
```
